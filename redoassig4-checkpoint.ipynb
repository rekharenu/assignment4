```python
# Dependencies and Setup
import pandas as pd
#from pathlib import path
# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv("schools_complete.csv")
student_data = pd.read_csv("students_complete.csv")

# Combine the data into a single dataset.

school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete
```

```python
schools_count = school_data_complete["school_name"].nunique()
student_count = school_data_complete["Student ID"].nunique()
total_budget = school_data_complete["budget"].sum()
print("Total number of schools:",schools_count)
print("Total number of students:",student_count)
print("Total budget:",total_budget)
```

```python
average_maths_score = school_data_complete['maths_score'].mean()
average_reading_score = school_data_complete['reading_score'].mean()
print("Average math score:",average_maths_score)
print("Average reading score:",average_reading_score)
```

```python
student_count= school_data_complete ["Student ID"].nunique()
passing_maths_count = school_data_complete [(school_data_complete["maths_score"] >= 50)].count()["Student ID"]
passing_maths_percentage = passing_maths_count / float(student_count) * 100
passing_reading_count = school_data_complete[(school_data_complete ["reading_score"]>=50)].count()["Student ID"]
passing_reading_percentage = passing_reading_count/student_count*100

passing_maths_reading_count = school_data_complete [(school_data_complete["maths_score"]>= 50) & (school_data_complete["reading_score"]>= 50)].count()["Student ID"]

overall_passing_rate = passing_maths_reading_count/student_count*100
print("Percentage of students passing math:",passing_maths_percentage) 
print("Percentage of students passing reading:",passing_reading_percentage) 
print("Percentage of students passing rate (student passing both maths and reading):",overall_passing_rate) 

```

```python
area_summary = pd.DataFrame({
    "Total Schools": [schools_count],
    "Total Students":[student_count],
    "Total Budget":[total_budget],
    "Average Math Score":[average_maths_score],
    "Average Reading Scaore":[average_reading_score],
    "% Passing Math":[passing_maths_percentage],
    "% passing Reading": [passing_reading_percentage],
    "% overall Passing Rate":[overall_passing_rate]
})

# Formatting
area_summary["Total Students"] = area_summary["Total Students"].map("{:,}".format)
area_summary["Total Budget"] = area_summary["Total Budget"].map("${:,.2f}".format)

# Display the DataFrame
print(area_summary)
```
